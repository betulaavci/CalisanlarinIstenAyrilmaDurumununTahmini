```python
from sklearn.preprocessing import LabelEncoder
```

```python
# LabelEncoder nesnesi oluşturulması
le = LabelEncoder()
# 'attrition' sütunundaki kategorik değerler sayısal değerlere dönüştürülmesi ve yeni bir 'label' sütununa eklenilmesi
df["label"] = le.fit_transform(df["attrition"])
# 'label' sütunundaki değerlerin veri tipi tam sayıya (int) dönüştürülmesi
df["label"] = df["label"].astype(int)
```

```python
# 'overtime' sütunundaki 'Yes' ve 'No' değerleri, 1 ve 0 ile değiştirilmesi
df["overtime"] = df["overtime"].replace({'Yes': 1, 'No': 0})

# 'gender' sütunundaki 'Male' ve 'Female' değerleri, 1 ve 0 ile değiştirilmesi
# 'inplace=True' ile değişiklikler doğrudan 'gender' sütununa uygulanılması
df["gender"] = df["gender"].replace({'Male': 1, 'Female': 0}, inplace=True)
```

```python
# DataFrame'in sütun isimlerini listelenmesi
df.columns
```

```python
# Bağımsız değişkenler (features) listesi oluşturulması
bagimsiz_degiskenler = ['age', 'businesstravel', 'dailyrate', 'department',
       'distancefromhome', 'education', 'educationfield',
       'environmentsatisfaction', 'gender', 'hourlyrate', 'jobinvolvement',
       'joblevel', 'jobrole', 'jobsatisfaction', 'maritalstatus',
       'monthlyincome', 'monthlyrate', 'numcompaniesworked', 'overtime',
       'percentsalaryhike', 'performancerating', 'relationshipsatisfaction',
       'stockoptionlevel', 'totalworkingyears', 'trainingtimeslastyear',
       'worklifebalance', 'yearsatcompany', 'yearsincurrentrole',
       'yearssincelastpromotion', 'yearswithcurrmanager']
```

```python
# 'bagimsiz_degiskenler' listesinde belirtilen sütunlardan değerler alınarak yeni bir 'features' sütunu oluşturulması
df["features"] = list(df[bagimsiz_degiskenler].values)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
# Bağımsız değişkenlerin (features) değerlerini X değişkenine atama
X = df[bagimsiz_degiskenler].values

# Bağımlı değişkenin (label) değerlerini y değişkenine atama
y = df["label"].values
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Her kodlama türü için sütunlar tanımlanıyor
categorical_features_ordinal = ['businesstravel', 'education', 'environmentsatisfaction', 'jobinvolvement',
                                 'jobsatisfaction', 'worklifebalance', 'performancerating', 'relationshipsatisfaction']
categorical_features_onehot = ['department', 'educationfield', 'jobrole', 'maritalstatus']
numerical_features = [col for col in bagimsiz_degiskenler if col not in categorical_features_ordinal + categorical_features_onehot]

# Her bir özellik türü için ön işleme boru hattı oluşturun.
#Sayısal boru hattına,NaN değerlerini ortalama ile doldurmak için SimpleImputer ekleyin."
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean'))  # Add this line
])

ordinal_pipeline = Pipeline([
    ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),  # Handle unknown values in OrdinalEncoder
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Add this line
])

onehot_pipeline = Pipeline([
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore')),
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

# Boru hatlarını ColumnTransformer kullanarak birleştirilmesi
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('ord', ordinal_pipeline, categorical_features_ordinal),
        ('ohe', onehot_pipeline, categorical_features_onehot)
    ])

# Ön işlemeyi eğitim ve test verilerine uygulayın.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Eğitim verisi üzerinde ön işleme adımlarını uygulamak
X_train = preprocessor.fit_transform(pd.DataFrame(X_train, columns=bagimsiz_degiskenler))

# Test verisi üzerinde ön işleme adımlarını uygulamak
X_test = preprocessor.transform(pd.DataFrame(X_test, columns=bagimsiz_degiskenler))
```

```python
# GradientBoostingClassifier modelini oluşturma
gbm = GradientBoostingClassifier(n_estimators=10)
```

```python
# GradientBoostingClassifier modelini eğitim verisi ile eğitme
gbm.fit(X_train, y_train)
```

```python
# Eğitimli GradientBoostingClassifier modeli ile test verisi üzerinde tahmin yapma
y_pred = gbm.predict(X_test)
```

```python
# Gerçek test etiketleri (y_test) ile tahmin edilen etiketler (y_pred) arasındaki doğruluğu hesaplama
accuracy = accuracy_score(y_test, y_pred)

# Modelin doğruluğunu yüzdelik formatta yazdırma
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Her kodlama türü için sütunları tanımlama
categorical_features_ordinal = ['businesstravel', 'education', 'environmentsatisfaction', 'jobinvolvement',
                                 'jobsatisfaction', 'worklifebalance', 'performancerating', 'relationshipsatisfaction']
categorical_features_onehot = ['department', 'educationfield', 'jobrole', 'maritalstatus']
numerical_features = [col for col in bagimsiz_degiskenler if col not in categorical_features_ordinal + categorical_features_onehot]

# Her özellik türü için ön işleme boru hattı oluşturma
# Sayısal boru hattına NaN değerlerini ortalama ile doldurmak için SimpleImputer ekleme
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean'))
])

# Sıralı (ordinal) kategorik veriler için boru hattı
ordinal_pipeline = Pipeline([
    ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),  # Sıralı kodlayıcıda bilinmeyen değerleri işleme
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Eksik değerleri en sık görülen değerle doldurma
])

# One-hot encoding kategorik veriler için boru hattı
onehot_pipeline = Pipeline([
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore')),  # Bilinmeyen değerleri yoksayma
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Eksik değerleri en sık görülen değerle doldurma
])

# Boru hatlarını ColumnTransformer ile birleştirme
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),  # Sayısal veriler için boru hattı
        ('ord', ordinal_pipeline, categorical_features_ordinal),  # Sıralı kategorik veriler için boru hattı
        ('ohe', onehot_pipeline, categorical_features_onehot)  # One-hot kategorik veriler için boru hattı
    ])

# Modeli eğitmeden önce veriye ön işleme uygulama
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train = preprocessor.fit_transform(pd.DataFrame(X_train, columns=bagimsiz_degiskenler))
X_test = preprocessor.transform(pd.DataFrame(X_test, columns=bagimsiz_degiskenler))

# Modeli oluşturma
gbm = GradientBoostingClassifier()

# Parametre ızgarasını oluşturma
param_grid = {
    'max_depth': [2, 4, 6],
    'max_features': [0.2, 0.5, 1.0],
    'n_estimators': [10, 20]
}

# GridSearchCV ile çapraz doğrulama ve hiperparametre optimizasyonu
grid_search = GridSearchCV(estimator=gbm, param_grid=param_grid, cv=10, scoring='accuracy', n_jobs=-1)

# Modeli ön işlenmiş veriye uygun şekilde eğitme
grid_search.fit(X_train, y_train)

# En iyi parametreleri ve en iyi skoru yazdırma
print("En iyi parametreler:", grid_search.best_params_)
print("En iyi skor:", grid_search.best_score_)

# Test seti üzerinde tahmin yapma
y_pred = grid_search.predict(X_test)

# Doğruluğu hesaplama
accuracy = accuracy_score(y_test, y_pred)
print("Test Doğruluğu:", accuracy)
```
