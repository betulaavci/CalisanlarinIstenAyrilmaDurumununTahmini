```python
from sklearn.preprocessing import LabelEncoder
```

```python
# LabelEncoder nesnesi oluÅturulmasÄ±
le = LabelEncoder()
# 'attrition' sÃ¼tunundaki kategorik deÄerler sayÄ±sal deÄerlere dÃ¶nÃ¼ÅtÃ¼rÃ¼lmesi ve yeni bir 'label' sÃ¼tununa eklenilmesi
df["label"] = le.fit_transform(df["attrition"])
# 'label' sÃ¼tunundaki deÄerlerin veri tipi tam sayÄ±ya (int) dÃ¶nÃ¼ÅtÃ¼rÃ¼lmesi
df["label"] = df["label"].astype(int)
```

```python
# 'overtime' sÃ¼tunundaki 'Yes' ve 'No' deÄerleri, 1 ve 0 ile deÄiÅtirilmesi
df["overtime"] = df["overtime"].replace({'Yes': 1, 'No': 0})

# 'gender' sÃ¼tunundaki 'Male' ve 'Female' deÄerleri, 1 ve 0 ile deÄiÅtirilmesi
# 'inplace=True' ile deÄiÅiklikler doÄrudan 'gender' sÃ¼tununa uygulanÄ±lmasÄ±
df["gender"] = df["gender"].replace({'Male': 1, 'Female': 0}, inplace=True)
```

```python
# BaÄÄ±msÄ±z deÄiÅkenler (features) listesi oluÅturulmasÄ±
bagimsiz_degiskenler = ['age', 'businesstravel', 'dailyrate', 'department',
       'distancefromhome', 'education', 'educationfield',
       'environmentsatisfaction', 'gender', 'hourlyrate', 'jobinvolvement',
       'joblevel', 'jobrole', 'jobsatisfaction', 'maritalstatus',
       'monthlyincome', 'monthlyrate', 'numcompaniesworked', 'overtime',
       'percentsalaryhike', 'performancerating', 'relationshipsatisfaction',
       'stockoptionlevel', 'totalworkingyears', 'trainingtimeslastyear',
       'worklifebalance', 'yearsatcompany', 'yearsincurrentrole',
       'yearssincelastpromotion', 'yearswithcurrmanager']
```

```python
# 'bagimsiz_degiskenler' listesinde belirtilen sÃ¼tunlardan deÄerler alÄ±narak yeni bir 'features' sÃ¼tunu oluÅturulmasÄ±
df["features"] = list(df[bagimsiz_degiskenler].values)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
# BaÄÄ±msÄ±z deÄiÅkenlerin (features) deÄerlerini X deÄiÅkenine atama
X = df[bagimsiz_degiskenler].values

# BaÄÄ±mlÄ± deÄiÅkenin (label) deÄerlerini y deÄiÅkenine atama
y = df["label"].values
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Her kodlama tÃ¼rÃ¼ iÃ§in sÃ¼tunlar tanÄ±mlanÄ±yor
categorical_features_ordinal = ['businesstravel', 'education', 'environmentsatisfaction', 'jobinvolvement',
                                 'jobsatisfaction', 'worklifebalance', 'performancerating', 'relationshipsatisfaction']
categorical_features_onehot = ['department', 'educationfield', 'jobrole', 'maritalstatus']
numerical_features = [col for col in bagimsiz_degiskenler if col not in categorical_features_ordinal + categorical_features_onehot]

# Her bir Ã¶zellik tÃ¼rÃ¼ iÃ§in Ã¶n iÅleme boru hattÄ± oluÅturun.
#SayÄ±sal boru hattÄ±na,NaN deÄerlerini ortalama ile doldurmak iÃ§in SimpleImputer ekleyin."
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean'))  # Add this line
])

ordinal_pipeline = Pipeline([
    ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),  # Handle unknown values in OrdinalEncoder
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Add this line
])

onehot_pipeline = Pipeline([
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore')),
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

# Boru hatlarÄ±nÄ± ColumnTransformer kullanarak birleÅtirilmesi
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('ord', ordinal_pipeline, categorical_features_ordinal),
        ('ohe', onehot_pipeline, categorical_features_onehot)
    ])

# Ãn iÅlemeyi eÄitim ve test verilerine uygulayÄ±n.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# EÄitim verisi Ã¼zerinde Ã¶n iÅleme adÄ±mlarÄ±nÄ± uygulamak
X_train = preprocessor.fit_transform(pd.DataFrame(X_train, columns=bagimsiz_degiskenler))

# Test verisi Ã¼zerinde Ã¶n iÅleme adÄ±mlarÄ±nÄ± uygulamak
X_test = preprocessor.transform(pd.DataFrame(X_test, columns=bagimsiz_degiskenler))
```

```python
# GradientBoostingClassifier modelini oluÅturma
gbm = GradientBoostingClassifier(n_estimators=10)
```

```python
# GradientBoostingClassifier modelini eÄitim verisi ile eÄitme
gbm.fit(X_train, y_train)
```

```python
# EÄitimli GradientBoostingClassifier modeli ile test verisi Ã¼zerinde tahmin yapma
y_pred = gbm.predict(X_test)
```

```python
# GerÃ§ek test etiketleri (y_test) ile tahmin edilen etiketler (y_pred) arasÄ±ndaki doÄruluÄu hesaplama
accuracy = accuracy_score(y_test, y_pred)

# Modelin doÄruluÄunu yÃ¼zdelik formatta yazdÄ±rma
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Her kodlama tÃ¼rÃ¼ iÃ§in sÃ¼tunlarÄ± tanÄ±mlama
categorical_features_ordinal = ['businesstravel', 'education', 'environmentsatisfaction', 'jobinvolvement',
                                 'jobsatisfaction', 'worklifebalance', 'performancerating', 'relationshipsatisfaction']
categorical_features_onehot = ['department', 'educationfield', 'jobrole', 'maritalstatus']
numerical_features = [col for col in bagimsiz_degiskenler if col not in categorical_features_ordinal + categorical_features_onehot]

# Her Ã¶zellik tÃ¼rÃ¼ iÃ§in Ã¶n iÅleme boru hattÄ± oluÅturma
# SayÄ±sal boru hattÄ±na NaN deÄerlerini ortalama ile doldurmak iÃ§in SimpleImputer ekleme
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean'))
])

# SÄ±ralÄ± (ordinal) kategorik veriler iÃ§in boru hattÄ±
ordinal_pipeline = Pipeline([
    ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),  # SÄ±ralÄ± kodlayÄ±cÄ±da bilinmeyen deÄerleri iÅleme
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Eksik deÄerleri en sÄ±k gÃ¶rÃ¼len deÄerle doldurma
])

# One-hot encoding kategorik veriler iÃ§in boru hattÄ±
onehot_pipeline = Pipeline([
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore')),  # Bilinmeyen deÄerleri yoksayma
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Eksik deÄerleri en sÄ±k gÃ¶rÃ¼len deÄerle doldurma
])

# Boru hatlarÄ±nÄ± ColumnTransformer ile birleÅtirme
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),  # SayÄ±sal veriler iÃ§in boru hattÄ±
        ('ord', ordinal_pipeline, categorical_features_ordinal),  # SÄ±ralÄ± kategorik veriler iÃ§in boru hattÄ±
        ('ohe', onehot_pipeline, categorical_features_onehot)  # One-hot kategorik veriler iÃ§in boru hattÄ±
    ])

# Modeli eÄitmeden Ã¶nce veriye Ã¶n iÅleme uygulama
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train = preprocessor.fit_transform(pd.DataFrame(X_train, columns=bagimsiz_degiskenler))
X_test = preprocessor.transform(pd.DataFrame(X_test, columns=bagimsiz_degiskenler))

# Modeli oluÅturma
gbm = GradientBoostingClassifier()

# Parametre Ä±zgarasÄ±nÄ± oluÅturma
param_grid = {
    'max_depth': [2, 4, 6],
    'max_features': [0.2, 0.5, 1.0],
    'n_estimators': [10, 20]
}

# GridSearchCV ile Ã§apraz doÄrulama ve hiperparametre optimizasyonu
grid_search = GridSearchCV(estimator=gbm, param_grid=param_grid, cv=10, scoring='accuracy', n_jobs=-1)

# Modeli Ã¶n iÅlenmiÅ veriye uygun Åekilde eÄitme
grid_search.fit(X_train, y_train)

# En iyi parametreleri ve en iyi skoru yazdÄ±rma
print("En iyi parametreler:", grid_search.best_params_)
print("En iyi skor:", grid_search.best_score_)

# Test seti Ã¼zerinde tahmin yapma
y_pred = grid_search.predict(X_test)

# DoÄruluÄu hesaplama
accuracy = accuracy_score(y_test, y_pred)
print("Test DoÄruluÄu:", accuracy)
```
